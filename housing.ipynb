```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load dataset from local path
housing = pd.read_csv(r"C:\Users\KIIT0001\Desktop\Oasis Infobyte Data Analytics\Housing\housing.csv")

# Outlier Analysis using boxplots for relevant numerical columns
fig, axs = plt.subplots(2, 3, figsize=(18, 10))

sns.boxplot(x=housing['Price'], ax=axs[0, 0])
axs[0, 0].set_title("Price")

sns.boxplot(x=housing['Avg. Area Income'], ax=axs[0, 1])
axs[0, 1].set_title("Avg. Area Income")

sns.boxplot(x=housing['Avg. Area House Age'], ax=axs[0, 2])
axs[0, 2].set_title("Avg. Area House Age")

sns.boxplot(x=housing['Avg. Area Number of Rooms'], ax=axs[1, 0])
axs[1, 0].set_title("Avg. Area Number of Rooms")

sns.boxplot(x=housing['Avg. Area Number of Bedrooms'], ax=axs[1, 1])
axs[1, 1].set_title("Avg. Area Number of Bedrooms")

sns.boxplot(x=housing['Area Population'], ax=axs[1, 2])
axs[1, 2].set_title("Area Population")

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

# Outlier Treatment for 'Price'
plt.boxplot(housing['Price'])
plt.title("Boxplot of Price (Before Outlier Removal)")
plt.show()

# Calculate IQR
Q1 = housing['Price'].quantile(0.25)
Q3 = housing['Price'].quantile(0.75)
IQR = Q3 - Q1

# Remove outliers
housing = housing[(housing['Price'] >= Q1 - 1.5 * IQR) & (housing['Price'] <= Q3 + 1.5 * IQR)]

# Confirm with a boxplot after treatment
plt.boxplot(housing['Price'])
plt.title("Boxplot of Price (After Outlier Removal)")
plt.show()
```

```python
import matplotlib.pyplot as plt

# Outlier Treatment for 'Area Population'
plt.boxplot(housing['Area Population'])
plt.title("Boxplot of Area Population (Before Outlier Removal)")
plt.show()

Q1 = housing['Area Population'].quantile(0.25)
Q3 = housing['Area Population'].quantile(0.75)
IQR = Q3 - Q1

# Remove outliers
housing = housing[(housing['Area Population'] >= Q1 - 1.5 * IQR) & 
                  (housing['Area Population'] <= Q3 + 1.5 * IQR)]

# Confirm with a boxplot
plt.boxplot(housing['Area Population'])
plt.title("Boxplot of Area Population (After Outlier Removal)")
plt.show()
```
